```python
#data pull removing retweets
num_of_tweets = 900
tweets = []
likes = []
time = []

for i in tweepy.Cursor(api.search, q="$GME -filter:retweets", tweet_mode ="extended", until = '2021-01-22').items(num_of_tweets):
    tweets.append(i.full_text)
    likes.append(i.favorite_count)
    time.append(i.created_at)
    
wsb_df = pd.DataFrame({
    'tweets': tweets
    ,'likes': likes
    ,'time': time
})
wsb_df.to_csv('twitter_data_1-22.csv')
```
